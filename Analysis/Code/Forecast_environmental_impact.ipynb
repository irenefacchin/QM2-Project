```
import matplotlib.pyplot as plt
import pandas as pd
import pylab

!curl https://raw.githubusercontent.com/irenefacchin/QM2-Project/main/Analysis/Code/Environmental%20impact%20fast%20fashion%20vs%20sustainable%20fashion%20(2015)%20v2.csv

wu_path = "https://raw.githubusercontent.com/irenefacchin/QM2-Project/main/Analysis/Code/Environmental%20impact%20fast%20fashion%20vs%20sustainable%20fashion%20(2015)%20v2.csv"
wu = pd.read_csv(wu_path)
wu.head(47)
```

```
import matplotlib.pyplot as plt
import pandas as pd
import pylab

wu = pd.read_csv(wu_path)

x = wu['Year']
y1 = wu['Fast Fashion CO2 Emissions']
plt.xlabel('Year') 
plt.ylabel('CO2 Emissions(g)')   
plt.plot(x,y1)
plt.title('CO2 Emissions Forecast Fast Fashion vs Sustainable Fashion')

y2 = wu['Sustainable Fashion CO2 Emissions']
plt.plot(x,y2)

plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
import pylab

wu = pd.read_csv(wu_path)

x = wu['Year']
y1 = wu['Fast Fashion Water Usage']
plt.xlabel('Year') 
plt.ylabel('Water Usage(L)')   
plt.plot(x,y1)
plt.title('Water Usage Forecast Fast Fashion vs Sustainable Fashion')

y2 = wu['Sustainable Fashion Water Usage']
plt.plot(x,y2)

plt.show()
```
